```python
### Name: Craig Dacosta
### Web Scape and Data Analysis

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
### Scape data from Wikipedia webste showing the largest companies by revenue

url = 'https://en.wikipedia.org/wiki/List_of_largest_companies_by_revenue'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
### find table
table = soup.find_all('table')[0]
```

```python
### find tr tag
soup.find('tr')
```

```python
### assign tr tag
world_companies = soup.find('tr')
```

world_companies

```python
### loop
world_table_companies = [ title.text.strip() for title in world_companies ]

print(world_table_companies)
```

```python
### print columns
df = pd.DataFrame(columns = world_table_companies)
df
```

```python
table.find_all('tr')
```

```python
column_data = table.find_all('tr')
```

```python
### loop and print dataframe
world_companies = [e.get_text(strip=True) for e in soup.find('tr')]
df = pd.DataFrame(columns = world_companies)

column_data = table.find_all('tr')
for row in column_data [2:]:
    individual_row_data = [data.text.strip() for data in row]
    length = len(df)
    df.loc[length] = individual_row_data
```

```python
### save csv file
df.to_csv('Lg_company_rev.csv')
```

```python
### Data Visualization using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### Q-1---Use scatter plot to answer question " Does the size of a company (number of employees) correlates with its revenue"?
sns.scatterplot(x='Employees', y='Revenue', hue='Name',style='Name', data=df, s=50)

plt.title('Scatter Plot of Company Size (Employees) vs. Revenue')
plt.xlabel('Number of Employees')
plt.ylabel('Revenue (in $ millions)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
```

```python
### Q-2--use bar chart to determine "Which industry generates the highest total revenue"?
plt.figure(figsize=(14, 10))
sns.set(rc={    
    "xtick.labelsize": 15,   
    "ytick.labelsize": 15    
})
sns.barplot(x='Industry', y='Revenue', data=df, hue='Industry', palette='viridis')

plt.title('Total Revenue by Industry')
plt.ylabel('Revenue (in $ millions)')
plt.xticks(rotation=70)
```

```python
### Q-3--Which company is most profitable after all expenses are paid from revenue?
plt.figure(figsize=(18, 10))
sns.set(rc={
    "axes.labelsize": 20,    
    "axes.titlesize": 20,    
    "xtick.labelsize": 17,   
    "ytick.labelsize": 17    
})
sns.barplot(x='Name', y='Profit', color='green', data=df)
plt.title('Most Profitable Companies', color='green')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
### Q-4--Company with the largest revenue?

plt.figure(figsize=(18, 10))
sns.set(rc={
    "axes.labelsize": 20,    
    "axes.titlesize": 20,    
    "xtick.labelsize": 17,   
    "ytick.labelsize": 17    
})
sns.barplot(x='Name', y='Revenue', color='Orange', data=df)
plt.title('Largest Revenue', color='Orange')
plt.xticks(rotation=90)
plt.tight_layout()
```

```python
### Q-5--"Which country, based on company headquarters, leads in terms of total revenue, profit, and workforce size?"
sns.pairplot(df, hue='Headquarters', markers=['o', 's', '^', 'D'], palette='Set1')
```
